# ***Django***

## ***Preparando o ambiente***

* `Criar ambiente virtual e ativar:` python -m venv nome_do_ambiente
* `Instalar django:` pip install django

## ***Criando o projeto***

primeiro comando é para iniciar o projeto o comando que devemos eexcultar é
```bash
django-admin startprojetc nome_do_projeto .
```
`atenção para o ponto no final do comando:` como o ponto no final "informamos" que queremos criar o prejeto com as configurações na raiz da pasta do projeto, como por exemplo na arquitetura abaixo:

```markdown
diretorio_atual/
├── MeuProjetoDjango/
│   ├── __init__.py
│   ├── asgi.py
│   ├── settings.py
│   ├── urls.py
│   └── wsgi.py
└── manage.py
```


`Sem o ponto no final do comando:` para as configurações será criada mais uma pasta (uma pasta dentro de outra pasta),  como por exemplo na arquitetura abaixo:
```markdown
diretorio_atual/
└── MeuProjetoDjango/
    ├── MeuProjetoDjango/
    │   ├── __init__.py
    │   ├── asgi.py
    │   ├── settings.py
    │   ├── urls.py
    │   └── wsgi.py
    └── manage.py
```
