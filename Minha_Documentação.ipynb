# ***Django***
O Django `é um framework web Python` de alto nível que incentiva o desenvolvimento rápido e um design limpo e pragmático. Construído por desenvolvedores experientes, ele cuida de grande parte do incômodo do desenvolvimento web, para que você possa se concentrar ao escrever seu aplicativo sem precisar reinventar a roda. Itilits livre e de código aberto.

## ***Preparando o ambiente***

* `Criar ambiente virtual e ativar:` python -m venv nome_do_ambiente
* `Instalar django:` pip install django

## ***Criando o projeto***

primeiro comando é para iniciar o projeto o comando que devemos eexcultar é
```bash
            django-admin startprojetc nome_do_projeto .
```
`atenção para o ponto no final do comando:` como o ponto no final "informamos" que queremos criar o prejeto com as configurações na raiz da pasta do projeto, como por exemplo na arquitetura abaixo:

```markdown
diretorio_atual/
├── MeuProjetoDjango/
│   ├── __init__.py
│   ├── asgi.py
│   ├── settings.py
│   ├── urls.py
│   └── wsgi.py
└── manage.py
```


`Sem o ponto no final do comando:` para as configurações será criada mais uma pasta (uma pasta dentro de outra pasta),  como por exemplo na arquitetura abaixo:
```markdown
diretorio_atual/
└── MeuProjetoDjango/
    ├── MeuProjetoDjango/
    │   ├── __init__.py
    │   ├── asgi.py
    │   ├── settings.py
    │   ├── urls.py
    │   └── wsgi.py
    └── manage.py
```

## ***Comandos django***
Digitando comando abaixo será listado os subcomando que podemos usar
```bash
 django-admin --help
```

esta tabela em ordem de uso mais necessário dos `subcomandos` usados pela comunidade Django:

`informação importante`
para os comando vamos usar 
```bash 
python manager.py
```
pois execulta a mesma coisa  que django-admin


| Subcomando          | Descrição                                                                 |
|---------------------|---------------------------------------------------------------------------|
| runserver           | Inicia o servidor de desenvolvimento do Django.                            |
| migrate             | Aplica e desaplica migrações no banco de dados.                            |
| makemigrations      | Cria novas migrações baseadas nas mudanças detectadas nos modelos do projeto. |
| startapp            | Cria uma nova aplicação Django em um diretório especificado.               |
| startproject        | Cria um novo projeto Django em um diretório especificado.                  |
| shell               | Abre o shell interativo do Django.                                         |
| createsuperuser     | Cria um novo superusuário (usuário administrativo).                        |
| dumpdata            | Serializa dados do banco de dados para um formato específico (como JSON).  |
| loaddata            | Carrega dados em formato JSON, XML, ou YAML para o banco de dados.         |
| showmigrations      | Mostra todas as migrações e indica quais foram aplicadas.                  |
| sqlmigrate          | Mostra o SQL para uma migração específica.                                 |
| test                | Executa os testes automatizados do projeto.                                |
| check               | Verifica se há problemas no projeto sem fazer migrações ou tocar no banco de dados. |
| inspectdb           | Inspeciona o banco de dados e gera modelos do Django baseados em suas tabelas. |
| flush               | Apaga todos os dados do banco de dados, mantendo as tabelas e suas estruturas. |
| compilemessages     | Compila arquivos de mensagens (.po) em arquivos de mensagens compilados (.mo). |
| makemessages        | Examina o código em busca de strings traduzíveis e cria arquivos de mensagens. |
| dbshell             | Abre o shell do banco de dados configurado para o seu projeto.             |
| sqlflush            | Mostra o SQL que será executado para apagar todos os dados do banco de dados. |
| sqlsequencereset    | Mostra o SQL para redefinir as sequências das tabelas indicadas.           |
| squashmigrations    | Mescla várias migrações em uma única migração.                             |
| diffsettings        | Mostra quais configurações foram alteradas do padrão do Django.            |
| createcachetable    | Cria a tabela de cache necessária se você estiver usando o cache baseado em banco de dados. |
| optimizemigration   | Otimiza os arquivos de migração existentes para reduzir a redundância.     |
| sendtestemail       | Envia um email de teste para verificar a configuração de email.            |
| testserver          | Inicia um servidor de desenvolvimento com dados de teste.                  |


Digitando o comando abaix, podemos  obter ajuda sobre um subcomando específico.
 ```bash 
 django-admin help <subcomando >
 ```


## ***Subindo o servidor - run server***


`OBS:` 
* 1º - é necessario estarmos no diretório do projeto, como o comando `pwd` no terminla é possivel saber em qual diretório esta 
* 2º - se esta usando um ambiente virtual `(venv)` e o django esta instalado la, se certificar que o ambiente virtual esta ativado 

### comando no terminal para subir o servidor é:


```bash
            python manage.py runserver
```
