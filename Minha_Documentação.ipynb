## ***Arquitetura de projeto Django***
entendendo melhor a arquitetura de um projeto Django
##### ***Uma Casa (Projeto Django)***

- `Projeto Django:` É a casa inteira. Ela é grande e tem várias salas e espaços diferentes.
- `Salas (Apps)`
- **App**: Cada app é como uma sala diferente da casa. Cada sala tem uma função específica. Por exemplo:
  - **Sala de Estar**: É onde você assiste TV e relaxa.
  - **Cozinha**: É onde você prepara e come a comida.
  - **Quarto**: É onde você dorme e guarda seus brinquedos.
- `Organização e Funções`
- Cada sala tem suas próprias coisas e faz algo diferente:
  - **Sala de Estar**: Tem sofás, TV, brinquedos.
  - **Cozinha**: Tem fogão, geladeira, panelas.
  - **Quarto**: Tem cama, guarda-roupa, brinquedos.
- `Fácil de Montar e Desmontar`
- Como cada sala tem suas próprias coisas, você pode montar e desmontar cada uma separadamente. Se você quiser usar a sala de estar em outra casa de brinquedo, é fácil de tirar e colocar em outra.

- `Resumo`
- Então, um **app** no Django é como uma dessas salas na sua casa de brinquedo. Ele faz uma coisa específica e tem tudo o que precisa para isso. Isso ajuda a manter tudo organizado e fácil de mexer.


### ***Path - definindo rotas***
detalhando a função path para a chamado das views

* `Um path` é como um caminho ou endereço que você segue para chegar a diferentes partes de um site.

A função `path` do módulo `django.urls` é usada para definir uma rota de URL no seu projeto Django. Aqui está um detalhamento dos argumentos que precisam ser passados para a função `path`:
```python
      def path(
          route: str,
          view: list[URLResolver | str],
          kwargs: dict[str, Any] = ...,
          name: str = ...,
      ) -> URLResolver: ...
```
### Argumentos da Função `path`

#### `route: str`

- **Descrição**: Este argumento define o padrão de URL como uma string. Pode incluir variáveis de rota, que são partes dinâmicas da URL.
- **Exemplo**: `'articles/<int:year>/'` define uma URL que inclui um número inteiro como parte da rota.

#### `view: IncludedURLConf`

- **Descrição**: Este argumento especifica a view que será chamada quando a URL corresponder ao padrão definido em `route`. Pode ser uma função de view, uma classe de view ou outra configuração de URL (via `include`).
- **Exemplo**:
  - **Function-based view**: `views.home` onde `views` é o módulo e `home` é a função.
  - **Class-based view**: `Home.as_view()` onde `Home` é uma classe de view.
  - **Incluindo outra configuração de URL**: `include('blog.urls')` onde `blog.urls` é um módulo de URLs.

#### `kwargs: dict[str, Any] = ...` (opcional)

- **Descrição**: Um dicionário de argumentos adicionais que podem ser passados para a view. Esse dicionário é usado para fornecer dados adicionais à view quando a URL é resolvida.
- **Exemplo**: `{'foo': 'bar'}` pode passar o argumento `foo` com valor `bar` para a view.

#### `name: str = ...` (opcional)

- **Descrição**: Um nome que você pode dar à rota de URL. Isso é útil para referenciar a URL em outras partes do seu código, especialmente nos templates.
- **Exemplo**: `name='home'` permite referenciar essa URL pelo nome `home`.

### Retorno da Função

#### `-> URLResolver`

- **Descrição**: A função `path` retorna um objeto `URLResolver`, que é usado internamente pelo Django para resolver a URL quando uma solicitação é recebida.


## ***Urls/rotas  aninhadas***

No Django, usamos a função include() para dividir nossas urls/rotas (ou arquivo urls.py) em `seções menores`, cada uma `controlada por um app diferente`:

`exemplo:` 
```python
include('app.urls'):
```
Isso significa que você está adicionando as URLs de um aplicativo específico ao mapa principal de URLs do seu projeto. Por exemplo, se você tiver um aplicativo que gerencia endpoints de uma API, você incluiria as URLs desse aplicativo no seu arquivo principal de URLs.

`Importação da Função include:` No arquivo urls.py do projeto principal (meuprojeto/urls.py), a função include é importada da biblioteca django.urls.
```python
            from django.urls import path, include
```

por padrão ao criar um APP, se olharmos não é criado no pacote (home) o modulo `urls.py`, então devemos criar e nele definir as urls do proprio APP 






dessa maneira simplificamos e não precisamos importar as `views dos APP`, importamos somente a `função ìnclude`, e nos parametros passamos 


exemplo de como fica as importações do arquivo de URLs principal do seu projeto
```python 
            from django.contrib import admin
            from django.urls import path, include

            urlpatterns = [
                # Inclui as URLs do app 'home' na raiz do site
                path('', include('home.urls')),      
                # Inclui as URLs do app 'results' para a rota 'results_home'
                path('results_home/', include('results.urls')),      
                # Rota para a administração do site
                path('admin/', admin.site.urls),
            ]

```

Dessa forma estamos chamando o modulo urls.py de cada APP, incluindo/concatenando as rotas que ha no modulo urls.py do proprio APPP

`Exemplo do modulo urls.py do APP RESULTS`

```python
            from django.urls import path
            # Importando as views do app 'results' e renomeando para 'views_results'
            from results import views as views_results 

            urlpatterns = [
                # Rota para a página inicial dos resultados
                path('', views_results.home),      
                # Rota para a página de clientes ativos
                path('clientes_ativos/', views_results.clientes_ativos),      
                # Rota para a página de clientes desativados
                path('clientes_desativados/', views_results.clientes_desativados),  
            ]

```



resumindo sempre será incluida/concatenada a rota principal as rotas do APP

## ***detalhando os parametros da função render:***

```python
    def render(
        request: HttpRequest,                # Parâmetro obrigatório: Representa o objeto HttpRequest, contendo informações sobre a requisição HTTP.
        template_name: str | Sequence[str],   # Parâmetro obrigatório: Nome ou lista de nomes de templates a serem renderizados.
        context: Mapping[str, Any] | None = ...,  # Parâmetro opcional: Dicionário contendo dados de contexto para o template.
        content_type: str | None = ...,       # Parâmetro opcional: Tipo de conteúdo HTTP da resposta.
        status: int | None = ...,             # Parâmetro opcional: Código de status HTTP da resposta.
        using: str | None = ...               # Parâmetro opcional: Nome do engine de template a ser utilizado.
    ) -> HttpResponse:                       # Tipo de retorno: Retorna um objeto HttpResponse.

```
então de acordo com o segundo parametro `templates_name` da função render, podemos passar o Nome ou lista de nomes de templates a que desejamos que sejam renderizados.

com isso podemos criar dentro de cada APP uma pasta `templates' 

### ***Name Space paa templates - Evitando colisão de nomes***
Como iremos lidar com varios templates, e por exemplo se queremos que cada APP tenha sua propria 'pagina home (exmplo index.html)', por padrã no django criamos `name space' com o nome do APP 

exemplo:

 


assim passamos para o paramentro `templates_name` da seguinte maneira

exemplo renderizando a `pagina home do APP 'home'`

```python
    def home(request): # aqui é a view home do APP home
        return render(request, 'home/index.html')
```


exemplo renderizando a `pagina home do APP 'results'`

## ***{%block %} e {%extends} - Extender TEMPLATES BASE (Herança)***

Resumindo é um template extendendo outro template (Herança)

A herança de templates ajuda a manter o código mais organizado e `evita a repetição de código`, facilitando a manutenção e a atualização do layout da aplicação. Aqui estão alguns termos específicos associados a essa prática:

* `Template Base:` O template que fornece a estrutura geral e que outros templates podem estender.
* `Template Filho:` O template que estende o template base e pode substituir partes específicas do conteúdo.
* `Blocos (Blocks):` Áreas nomeadas dentro do template base que podem ser substituídas ou preenchidas pelo conteúdo dos templates filhos.

### exemplo de uso 
Podemos criar um `thema padrão` e reutilizalos em outros templates, com ecarrega-lo e o templete atual poderá extender alguns blocos do template padrão 

### Instalar extensão django

para poder usar o django.html como linguagem devemos baixar a extensão que nos permitira usar  
```python
    {% extends  %}  e  {% block %}{% endblock %} 
```
`esta é a extensão que deve ser instalada` 



```python
    {% extends "TemplateBase.html" %}  # aqui será chamamos o templete html que será extendido

    {% block NomeDoBloco %}{% endblock NomeDoBloco %} # aqui é o bloco que será alterado no template atual
```
### ***Explicando***

#### ***{% extends  %}***
```python
    {% extends "TemplateBase.html" %}  # aqui será chamamos o templete html que será extendido
```
    O que faz?
    Essa tag é usada para indicar que o template atual herda de outro template, chamado "templateBase.html"



#### ***{% block %}{% endblock %}***
```python
    {% block NomeDoBloco %}{% endblock NomeDoBloco %} # aqui é o bloco que será alterado no template atual
```
O que faz?
Essa tag é usada para definir uma seção (bloco) de conteúdo dentro do template base que pode ser sobrescrita pelo template filho. Dentro do bloco, você coloca o conteúdo que deseja inserir ou modificar.

Aqui, `"NomeDoBloco"` é o nome do bloco. O conteúdo dentro dessas tags substitui o conteúdo do bloco correspondente no template base.
