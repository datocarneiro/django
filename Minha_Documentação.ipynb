# ***Django***
O Django `é um framework web Python` de alto nível que incentiva o desenvolvimento rápido e um design limpo e pragmático. Construído por desenvolvedores experientes, ele cuida de grande parte do incômodo do desenvolvimento web, para que você possa se concentrar ao escrever seu aplicativo sem precisar reinventar a roda. Itilits livre e de código aberto.

## ***Preparando o ambiente***

* `Criar ambiente virtual e ativar:` python -m venv nome_do_ambiente
* `Instalar django:` pip install django

## ***Criando o projeto***

primeiro comando é para iniciar o projeto o comando que devemos eexcultar é
```bash
            django-admin startprojetc nome_do_projeto .
```
`atenção para o ponto no final do comando:` como o ponto no final "informamos" que queremos criar o prejeto com as configurações na raiz da pasta do projeto, como por exemplo na arquitetura abaixo:

```markdown
diretorio_atual/
├── MeuProjetoDjango/
│   ├── __init__.py
│   ├── asgi.py
│   ├── settings.py
│   ├── urls.py
│   └── wsgi.py
└── manage.py
```


`Sem o ponto no final do comando:` para as configurações será criada mais uma pasta (uma pasta dentro de outra pasta),  como por exemplo na arquitetura abaixo:
```markdown
diretorio_atual/
└── MeuProjetoDjango/
    ├── MeuProjetoDjango/
    │   ├── __init__.py
    │   ├── asgi.py
    │   ├── settings.py
    │   ├── urls.py
    │   └── wsgi.py
    └── manage.py
```

## ***Arquitetura de projeto Django***
entendendo melhor a arquitetura de um projeto Django
##### ***Uma Casa (Projeto Django)***

- `Projeto Django:` É a casa inteira. Ela é grande e tem várias salas e espaços diferentes.
- `Salas (Apps)`
- **App**: Cada app é como uma sala diferente da casa. Cada sala tem uma função específica. Por exemplo:
  - **Sala de Estar**: É onde você assiste TV e relaxa.
  - **Cozinha**: É onde você prepara e come a comida.
  - **Quarto**: É onde você dorme e guarda seus brinquedos.
- `Organização e Funções`
- Cada sala tem suas próprias coisas e faz algo diferente:
  - **Sala de Estar**: Tem sofás, TV, brinquedos.
  - **Cozinha**: Tem fogão, geladeira, panelas.
  - **Quarto**: Tem cama, guarda-roupa, brinquedos.
- `Fácil de Montar e Desmontar`
- Como cada sala tem suas próprias coisas, você pode montar e desmontar cada uma separadamente. Se você quiser usar a sala de estar em outra casa de brinquedo, é fácil de tirar e colocar em outra.

- `Resumo`
- Então, um **app** no Django é como uma dessas salas na sua casa de brinquedo. Ele faz uma coisa específica e tem tudo o que precisa para isso. Isso ajuda a manter tudo organizado e fácil de mexer.

## ***View***
criando views no modulo views do app



```python
      from django.shortcuts import render
      from django.http import HttpResponse

      # Create your views here.
      def home(request):
          return HttpResponse('aqui é a HOME')
```
`obs:` 
  * estamos importando a função render, que se faz necesario para renderizar a view no navegador
  * estamos importanto tambem a classe HttpResponse para poder entregar uma response (resposta)



## ***Acessando as views***

na pasta do projeto no modulo `urls.py`precisamos importar o modulo views do app pretendido

`dica:` para a importação do modulo para não haver erros, sempre começe pela raiz do App nesse exemplo vamos fazer do APP home e APP results
ficaria assim:
```python
            from home import views as views_home    # com alias 
            from home import views as views_results    # com alias 
```
`obs:` para as views usar sempre um `alias (apelido)`, por que todos os APP que forem criado terão o modulo views.py isso garante que não haverá problemas ao acessar a views do app pretendido

## detalhando a funç~çao path para a chamado das views

A função `path` do módulo `django.urls` é usada para definir uma rota de URL no seu projeto Django. Aqui está um detalhamento dos argumentos que precisam ser passados para a função `path`:
```python
      def path(
          route: str,
          view: list[URLResolver | str],
          kwargs: dict[str, Any] = ...,
          name: str = ...,
      ) -> URLResolver: ...
```
### Argumentos da Função `path`

#### `route: str`

- **Descrição**: Este argumento define o padrão de URL como uma string. Pode incluir variáveis de rota, que são partes dinâmicas da URL.
- **Exemplo**: `'articles/<int:year>/'` define uma URL que inclui um número inteiro como parte da rota.

#### `view: IncludedURLConf`

- **Descrição**: Este argumento especifica a view que será chamada quando a URL corresponder ao padrão definido em `route`. Pode ser uma função de view, uma classe de view ou outra configuração de URL (via `include`).
- **Exemplo**:
  - **Function-based view**: `views.home` onde `views` é o módulo e `home` é a função.
  - **Class-based view**: `Home.as_view()` onde `Home` é uma classe de view.
  - **Incluindo outra configuração de URL**: `include('blog.urls')` onde `blog.urls` é um módulo de URLs.

#### `kwargs: dict[str, Any] = ...` (opcional)

- **Descrição**: Um dicionário de argumentos adicionais que podem ser passados para a view. Esse dicionário é usado para fornecer dados adicionais à view quando a URL é resolvida.
- **Exemplo**: `{'foo': 'bar'}` pode passar o argumento `foo` com valor `bar` para a view.

#### `name: str = ...` (opcional)

- **Descrição**: Um nome que você pode dar à rota de URL. Isso é útil para referenciar a URL em outras partes do seu código, especialmente nos templates.
- **Exemplo**: `name='home'` permite referenciar essa URL pelo nome `home`.

### Retorno da Função

#### `-> URLResolver`

- **Descrição**: A função `path` retorna um objeto `URLResolver`, que é usado internamente pelo Django para resolver a URL quando uma solicitação é recebida.
